```python
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import StandardScaler
```

```python
output_folder = './output/competition_synergy'
```

```python
df = pd.read_csv('data/Riva_del_garda_hotels.csv', sep=',', index_col=0)
df.drop(columns=['tot_revenue_euro', 'city','%_option_not_specified','%_option_all_inclusive'], inplace=True, errors='ignore')
df
```

```python
df.describe()
```

```python
df.columns
```

```python
# scale with StandardScaler
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)
```

```python
mat = df.T.corr('spearman').round(2)
mat
```

```python
# plot the correlation matrix with plotly
fig = px.imshow(mat, text_auto=True, aspect='auto', color_continuous_scale='RdBu_r', zmin=-1, zmax=1)
fig.update_layout(title='Competition-Synergy Matrix (Riva Del Garda)', height=800)
fig.show()
fig.write_image(f'{output_folder}/competition_synergy_matrix.png')
```

```python
# create a bar blot of the mean correlation values
mean_corr = mat.mean().sort_values(ascending=False).round(2)
fig = px.bar(mean_corr, x=mean_corr.index, y=mean_corr.values,
             labels={'x': 'Features', 'y': 'Mean Competition-Synergy'},
             title='Mean Competition-Synergy of Features (Riva Del Garda)',
             color=mean_corr.values, color_continuous_scale='RdBu_r', text_auto=True, range_color=[-1, 1])
fig.update_layout(height=600, xaxis_title='Features', yaxis_title='Mean Competition-Synergy')
fig.show()
fig.write_image(f'{output_folder}/mean_competition_synergy.png')
```

```python
# compare two random hotels with bar plot with plotly
hotel1 = df.sample(1, random_state=22).T
hotel1
```

```python
hotel2 = df.sample(1, random_state=42).T
hotel2
```

```python
comparison_df = pd.concat([hotel1, hotel2], axis=1).round(2)
comparison_df.columns = ['Hotel 24', 'Hotel 9']
fig = px.bar(comparison_df, text_auto=True,
             labels={'value': 'Scaled Value', 'variable': 'Hotels'},
             title='Comparison of Two Random Hotels (Riva Del Garda)',
             color_discrete_sequence=['#636EFA', "#3BEF71"], orientation='h')
fig.update_layout(height=1000, xaxis_title='Features', yaxis_title='Scaled Value')
fig.show()
fig.write_image(f'{output_folder}/comparison_hotels_example.png')
```
